Bu bitirme metotları, dövüşün nasıl sonuçlandığını gösterir ve UFC’de dövüş stratejilerini anlamak için önemli bir değişkendir.

**Decision - Unanimous:** Üç hakemin de aynı dövüşçüyü kazanan seçtiği hakem kararıdır.

**Decision - Split:** Hakemlerden ikisi bir dövüşçüyü, biri diğer dövüşçüyü kazanan seçtiğinde oluşur.

**Decision - Majority:** İki hakem bir dövüşçüyü kazanan seçer, bir hakem beraberlik verir.

**KO/TKO:** Rakibin nakavt olması ya da hakem/doktor tarafından dövüşe devam edemeyecek durumda görülmesidir.

**TKO - Doctor's Stoppage:** Doktorun sakatlık, kesik veya sağlık riski nedeniyle dövüşü durdurmasıdır.

**Submission:** Bir dövüşçünün boğma, kilit veya baskı sonucu pes etmesidir.

**DQ:** Kural ihlali nedeniyle diskalifiye ile sonuçlanan dövüşlerdir.

**Could Not Continue:** Dövüşçünün sakatlık veya benzeri bir nedenle devam edemediği durumlardır.

**Overturned:** Dövüş sonucunun sonradan resmi olarak değiştirilmesidir.

Bu tabloya göre UFC’de dövüşlerin önemli bir kısmı hakem kararına gitse de KO/TKO ve Submission gibi bitirici sonuçlar da oldukça yüksektir.

Dövüşlerin bitiş yöntemlerindeki kırılma noktalarını tespit ederek, sporun zaman içerisinde daha sert mi yoksa daha stratejik ve dengeli bir yapıya mı evrildiğini veriyle ortaya koymaktır.

```python
df_analiz = df_master.copy()

benzersiz_metotlar = df_analiz['method'].value_counts()
print(benzersiz_metotlar)
```

```python
eksik_veri = df_analiz['method'].isna().sum()
toplam_veri = len(df_analiz)
print(f"Toplam Satır: {toplam_veri} | Eksik (NaN) 'method' Sayısı: {eksik_veri}")
```

```python
import pandas as pd
import plotly.express as px

df_kumulatif = df_master.copy()
df_kumulatif['date'] = pd.to_datetime(df_kumulatif['date'])
df_kumulatif['Year'] = df_kumulatif['date'].dt.year

df_counts = df_kumulatif.groupby(['Year', 'method']).size().reset_index(name='Sayi')

tum_yillar = sorted(df_counts['Year'].unique())
tum_metotlar = df_counts['method'].unique()
index_ornek = pd.MultiIndex.from_product([tum_yillar, tum_metotlar], names=['Year', 'method'])
df_counts = df_counts.set_index(['Year', 'method']).reindex(index_ornek, fill_value=0).reset_index()

df_counts['Kumulatif_Sayi'] = df_counts.groupby('method')['Sayi'].cumsum()
df_counts = df_counts.sort_values('Year')

silinecek_metotlar = ['Overturned', 'Other', 'Could Not Continue', 'DQ', 'CNC']
df_gorsel = df_counts[~df_counts['method'].isin(silinecek_metotlar)].copy()

fig = px.bar(
    df_gorsel,
    x="method",
    y="Kumulatif_Sayi",
    color="method",
    animation_frame="Year",
    animation_group="method",
    range_y=[0, df_gorsel['Kumulatif_Sayi'].max() + 50],
    title="UFC Ana Dövüş Bitiş Yöntemlerinin Kümülatif Artışı (1994 - 2025)",
    labels={"method": "Bitiş Yöntemi", "Kumulatif_Sayi": "Toplam Birikimli Dövüş Sayısı", "Year": "Yıl"},
    color_discrete_sequence=px.colors.qualitative.Dark24
)

fig.update_layout(
    showlegend=True,
    xaxis_title="Ana Bitiş Yöntemleri",
    yaxis_title="Toplam Dövüş Sayısı",
    template="plotly_white",
    xaxis={'categoryorder':'total descending'}
)

fig.show()
```

Bulgular ve Grafik Yorumu

Yıllara göre olarak biriken animasyonlu grafik incelendiğinde, organizasyonun net kırılma noktaları ve evrimi şu şekilde özetlenmektedir:

1. **İlkel Dönem (1994 - 1998):** UFC'nin başlangıç yıllarında, sınırlı sayıdaki müsabakalarda **Submission (Pes Ettirme)** yönteminin belirgin bir ağırlığı olduğu görülmektedir. Sporun daha geniş kitlelere ulaşması, boks ve kickboks kökenli dövüşçülerin organizasyona dahil olmasıyla UFC, ayakta sert darbelerin ve nakavtların domine ettiği, fiziksel yıpratıcılığın en üst düzeye çıktığı bir spor haline gelmiştir.
3. **Modern Taktiksel Dönem (2015 - 2025):** 2015 sonrasında ise grafikte en çarpıcı kırılma gerçekleşmiş ve **U-DEC (Oy Birliği ile Karar)** dikey eksende dikey bir sıçrama yaparak tüm bitiş yöntemlerine çok ciddi bir fark atmıştır.

## 2. Mekansal Analiz: UFC'nin Küresel Genişleme Stratejisi ve Küreselleşme Haritası (1994 - 2025)

**Soru:** UFC etkinlikleri 1994–2025 yılları arasında dünya genelinde hangi ülkelerde yoğunlaşmıştır, hangi ülkeler organizasyonun küresel yayılımında öne çıkmıştır ve bu dağılım UFC’nin uluslararası spor pazarı içindeki konumunu nasıl yansıtmaktadır?

**Analizin Amacı:**  
Bu analizin amacı, UFC etkinliklerinin dünya genelindeki coğrafi dağılımını ülke bazında incelemektir. Etkinlik lokasyonları ayrıştırılarak her ülkede düzenlenen organizasyon sayısı, toplam dövüş sayısı, kemer maçı sayısı ve ev sahibi dövüşçü oranı analiz edilmiştir. Böylece UFC’nin hangi ülkelerde daha yoğun faaliyet gösterdiği ve küresel ölçekte hangi pazarların öne çıktığı görsel olarak değerlendirilmiştir.

Veri Kataloğu

**Veri Kümesi:** `df_events` ve dövüş detaylarından türetilen ülke bazlı özet veri

| Değişken | Açıklama | Veri Türü |
|---|---|---|
| `Ulke` | UFC etkinliğinin düzenlendiği ülkeyi gösterir. | Kategorik |
| `Düzenlenen Etkinlik Sayısı` | Her ülkede düzenlenen toplam UFC etkinliği sayısını gösterir. | Sayısal |
| `Toplam Yapılan Dövüş Sayısı` | Her ülkede gerçekleşen toplam dövüş sayısını gösterir. | Sayısal |
| `Düzenlenen Ünvan (Kemer) Maçı` | Her ülkede yapılan toplam şampiyonluk/kemer maçı sayısını gösterir. | Sayısal |
| `Ev Sahibi Dövüşçü Oranı` | O ülkedeki etkinliklerde yerel dövüşçülerin temsil oranını gösterir. | Sayısal |
| `Renk Yoğunluğu` | Haritadaki renk değişimi, ülkelerdeki etkinlik yoğunluğunu görsel olarak temsil eder. | Sayısal |

```python
import numpy as np
import pandas as pd
import plotly.express as px

df_geo = df_master.copy()

ulke_sozlugu = {
    "USA": "United States",
    "UK": "United Kingdom",
    "UAE": "United Arab Emirates",
    "South Korea": "Korea, Republic of",
    "Russia": "Russian Federation",
}

df_geo["Ulke"] = (
    df_geo["location"].astype(str).str.split(",").str[-1].str.strip().replace(ulke_sozlugu)
)

title_col = next(
    (c for c in df_geo.columns if any(k in c.lower() for k in ["title", "bout"])), None
)
df_geo["title_bout_numeric"] = (
    pd.to_numeric(df_geo[title_col], errors="coerce").fillna(0).astype(int)
    if title_col else 0
)

df_harita_veri = (
    df_geo.groupby("Ulke")
    .agg(
        Toplam_Dovus=("date", "size"),
        Etkinlik_Sayisi=("date", "nunique"),
        Kemer_Maclari=("title_bout_numeric", "sum"),
    )
    .reset_index()
)

df_harita_veri["Log_Etkinlik"] = np.log10(df_harita_veri["Etkinlik_Sayisi"] + 1)
df_harita_veri["Ev_Sahibi_Orani"] = (
    (df_harita_veri["Kemer_Maclari"] / df_harita_veri["Toplam_Dovus"] * 100)
    .round(1).astype(str) + "%"
)

fig = px.choropleth(
    df_harita_veri,
    locations="Ulke",
    locationmode="country names",
    color="Log_Etkinlik",
    hover_name="Ulke",
    hover_data={
        "Log_Etkinlik": False,
        "Etkinlik_Sayisi": True,
        "Toplam_Dovus": True,
        "Kemer_Maclari": True,
        "Ev_Sahibi_Orani": True,
    },
    title="UFC Küresel Genişleme ve Stratejik Ülke Analizi (1994 - 2025)",
    color_continuous_scale=px.colors.sequential.YlOrRd,
    labels={
        "Etkinlik_Sayisi": "Düzenlenen Etkinlik Sayısı",
        "Toplam_Dovus": "Toplam Yapılan Dövüş Sayısı",
        "Kemer_Maclari": "Düzenlenen Unvan (Kemer) Maçı",
        "Ev_Sahibi_Orani": "Ev Sahibi Dövüşçü Oranı",
    },
)

fig.update_layout(
    template="plotly_dark",
    coloraxis_showscale=False,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth",
        landcolor="#1a1a1a",
        oceancolor="#0a0a14",
        showocean=True,
        lakecolor="#0a0a14",
    ),
    margin={"r": 0, "t": 50, "l": 0, "b": 0},
)

fig.show()
```

```python
import pandas as pd
import numpy as np
import json
import plotly.offline as pyo
from IPython.display import display, HTML

pyo.init_notebook_mode(connected=False)

R_NAME = 'r_name'
B_NAME = 'b_name'

METRIKLER = {
    'Sig Str':  ('r_sig_str_landed', 'b_sig_str_landed'),
    'Kafa':     ('r_head_landed',    'b_head_landed'),
    'Gövde':    ('r_body_landed',    'b_body_landed'),
    'Bacak':    ('r_leg_landed',     'b_leg_landed'),
    'Takedown': ('r_td_landed',      'b_td_landed'),
    'KD':       ('r_kd',             'b_kd'),
    'Sub Att':  ('r_sub_att',        'b_sub_att'),
    'Str Def':  ('r_str_def',        'b_str_def'),
}

RENKLER_HEX  = ['#FF4444', '#4499FF', '#44FF99', '#FFD700', '#FF44FF', '#FF8C00']
RENKLER_RGBA = [
    'rgba(255,68,68,0.15)',  'rgba(68,153,255,0.15)',
    'rgba(68,255,153,0.15)', 'rgba(255,215,0,0.15)',
    'rgba(255,68,255,0.15)', 'rgba(255,140,0,0.15)',
]

def temizle_isim(isim):
    return str(isim).replace('-', ' ').title().strip()

GLOBAL_MAKS = {}
for etiket, (r_col, b_col) in METRIKLER.items():
    tum = pd.concat([
        pd.to_numeric(df_master[r_col], errors='coerce'),
        pd.to_numeric(df_master[b_col], errors='coerce'),
    ]).dropna()
    GLOBAL_MAKS[etiket] = float(tum.max()) if len(tum) > 0 else 1.0

etiketler = list(METRIKLER.keys())

all_means = {}
for etiket, (r_col, b_col) in METRIKLER.items():
    r_df = df_master[[R_NAME, r_col]].rename(columns={R_NAME: 'isim', r_col: 'val'})
    b_df = df_master[[B_NAME, b_col]].rename(columns={B_NAME: 'isim', b_col: 'val'})
    r_df['isim'] = r_df['isim'].apply(temizle_isim)
    b_df['isim'] = b_df['isim'].apply(temizle_isim)
    combined = pd.concat([r_df, b_df])
    combined['val'] = pd.to_numeric(combined['val'], errors='coerce')
    all_means[etiket] = combined.groupby('isim')['val'].mean()

tum_isimler = sorted(set().union(*[m.index for m in all_means.values()]))

dovuscu_data = {}
for isim in tum_isimler:
    norms, hams = [], []
    for etiket in etiketler:
        v = all_means[etiket].get(isim, np.nan)
        hams.append(0.0 if pd.isna(v) else round(float(v), 2))
        norms.append(0.0 if pd.isna(v) else round(float(v) / GLOBAL_MAKS[etiket] * 100, 1))
    dovuscu_data[isim] = {'n': norms, 'h': hams}

options_html = '\n'.join(f'<option value="{d}">{d}</option>' for d in tum_isimler)

html = f"""
<div style="background:#111;padding:15px;border-radius:8px;font-family:sans-serif;">
  <div style="display:flex;gap:8px;align-items:center;margin-bottom:10px;flex-wrap:wrap;">
    <input id="ara" type="text" placeholder="Dövüşçü ara..." oninput="filtrele()"
           style="padding:6px;border-radius:4px;background:#222;color:white;border:1px solid #444;width:180px;">
    <select id="sel"
            style="width:240px;padding:6px;border-radius:4px;background:#222;color:white;border:1px solid #444;">
      {options_html}
    </select>
    <button onclick="ekle()"
            style="background:#28a745;color:white;border:none;padding:6px 16px;border-radius:4px;cursor:pointer;font-size:14px;">Ekle</button>
    <button onclick="temizle()"
            style="background:#dc3545;color:white;border:none;padding:6px 16px;border-radius:4px;cursor:pointer;font-size:14px;">Temizle</button>
    <span id="msg" style="color:#f90;font-size:13px;"></span>
  </div>
  <div style="display:flex;align-items:center;gap:10px;margin-bottom:10px;">
    <span style="color:#aaa;font-size:13px;">Zoom:</span>
    <input id="zoom" type="range" min="10" max="100" value="100" step="5"
           oninput="zoomGuncelle(this.value)"
           style="width:300px;accent-color:#4499FF;">
    <span id="zoom-val" style="color:white;font-size:13px;min-width:40px;">100%</span>
  </div>
  <div id="eklenenler" style="margin-bottom:8px;min-height:20px;"></div>
  <div id="chart"></div>
</div>
<script>
(function() {{
  const DATA = {json.dumps(dovuscu_data)};
  const ET   = {json.dumps(etiketler)};
  const HEX  = {json.dumps(RENKLER_HEX)};
  const RGBA = {json.dumps(RENKLER_RGBA)};
  const allOpts = Array.from(document.getElementById('sel').options);
  let eklenen = [];
  let zoom = 100;

  function filtrele() {{
    const q = document.getElementById('ara').value.toLowerCase();
    allOpts.forEach(o => o.style.display = o.value.toLowerCase().includes(q) ? '' : 'none');
    const vis = allOpts.find(o => o.style.display !== 'none');
    if (vis) document.getElementById('sel').value = vis.value;
  }}

  function ekle() {{
    const isim = document.getElementById('sel').value;
    const msg  = document.getElementById('msg');
    if (!isim) return;
    if (eklenen.includes(isim)) {{ msg.textContent = isim + ' zaten eklendi.'; return; }}
    if (eklenen.length >= 6)   {{ msg.textContent = 'Maksimum 6 dövüşçü.'; return; }}
    eklenen.push(isim);
    msg.textContent = '';
    listele();
    ciz();
  }}

  function temizle() {{
    eklenen = [];
    document.getElementById('msg').textContent = '';
    document.getElementById('eklenenler').innerHTML = '';
    Plotly.purge('chart');
  }}

  function listele() {{
    document.getElementById('eklenenler').innerHTML =
      eklenen.map((isim, i) =>
        `<span style="color:${{HEX[i]}};margin-right:14px;font-size:13px;">● ${{isim}}</span>`
      ).join('');
  }}

  function zoomGuncelle(val) {{
    zoom = parseInt(val);
    document.getElementById('zoom-val').textContent = zoom + '%';
    if (eklenen.length > 0) ciz();
  }}

  function getLayout() {{
    const tickAralik = zoom / 5;
    const tickvals = [1,2,3,4,5].map(i => Math.round(tickAralik * i));
    return {{
      polar: {{
        bgcolor: '#141414',
        radialaxis: {{visible:true, range:[0,zoom], tickvals:tickvals, tickfont:{{color:'gray',size:9}}, gridcolor:'#333', linecolor:'#333'}},
        angularaxis: {{tickfont:{{color:'white',size:12}}, gridcolor:'#333', linecolor:'#333'}},
      }},
      paper_bgcolor: '#111111',
      font: {{color:'white'}},
      legend: {{bgcolor:'#222', bordercolor:'#444', font:{{color:'white',size:11}}}},
      width: 620, height: 600,
      margin: {{l:60,r:60,t:40,b:40}},
      hoverlabel: {{bgcolor:'#222', bordercolor:'#444', font:{{color:'white',size:12}}}},
    }};
  }}

  function ciz() {{
    const traces = eklenen.map((isim, i) => {{
      const d  = DATA[isim];
      const cd = ET.map((e, j) => `<b>${{e}}</b><br>Ort: ${{d.h[j].toFixed(2)}}<br>Skala: ${{d.n[j].toFixed(1)}}/100`);
      return {{
        type: 'scatterpolar',
        r: [...d.n, d.n[0]],
        theta: [...ET, ET[0]],
        fill: 'toself',
        fillcolor: RGBA[i],
        line: {{color: HEX[i], width: 2}},
        name: isim,
        customdata: [...cd, cd[0]],
        hovertemplate: '%{{customdata}}<extra>' + isim + '</extra>',
      }};
    }});
    Plotly.react('chart', traces, getLayout());
  }}

  window.ekle     = ekle;
  window.temizle  = temizle;
  window.filtrele = filtrele;
  window.zoomGuncelle = zoomGuncelle;

  if (typeof Plotly !== 'undefined') {{
  }} else if (typeof require !== 'undefined') {{
    require(['plotly'], function(P) {{ window.Plotly = P; }});
  }} else {{
    var s = document.createElement('script');
    s.src = 'https://cdn.plot.ly/plotly-2.26.0.min.js';
    document.head.appendChild(s);
  }}
}})();
</script>
"""

display(HTML(html))
```
